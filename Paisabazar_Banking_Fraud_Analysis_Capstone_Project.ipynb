# **Problem Statement**


The objective of this project is to build an automated machine learning classification model for Paisabazaar that accurately predicts a customer's credit score tier—categorized as Good, Standard, or Poor—by evaluating their multi-dimensional financial behavior, including annual income, outstanding debt, credit utilization ratio, and historical payment delays tracked over multiple months. In the banking sector, traditional manual credit underwriting is highly inefficient and elevates risk exposure, often leading to either high-risk loan defaults or the rejection of creditworthy clients. By developing this predictive framework, the institution can implement automated, real-time credit decisioning, which effectively minimizes Non-Performing Assets (NPAs), accelerates the loan and credit card approval lifecycle, and optimizes overall risk management.
